**Outlier Removal**

1. Using InterQuantile Range(IQR)

```python
l = [11,10,-38,12,14,19,20,-56,100,156,23,-97,56,89,189,-10]
#converting list in to dataframe
df = pd.DataFrame(l,columns = ["Data"]) #data is the name of the column
df
```

```python
df = df.sort_values(by='Data') # by = "which value is using for sorting"
df
```

```python
df.describe()
```

```python
q1 = df.quantile(.25)
q3 = df.quantile(0.75)
print(q1,"\n",q3)
```

```python
iqr = q3-q1
iqr
```

```python
lb = q1-(1.5*iqr)
ub = q3+(1.5*iqr)
print(lb,"\n",ub) # whaterver in this range is considered and others are outliers
```

```python
df[(df<lb)]
```

```python
df[(df<ub)]
```

```python
df_new = df[(df>lb)&(df<ub)]
df_new
```

```python
df.describe()
```

```python
df_new.describe()
```

2. Using SD

```python
l = [11,10,-38,12,14,19,260,-566,100,156,23,-97,56,89,189,-10]
#converting list in to dataframe
df = pd.DataFrame(l,columns = ["Data"]) #data is the name of the column
df
```

```python
df = df.sort_values(by = "Data") # by = "which value is using for sorting"
df
```

```python
ub = df.Data.mean()+3*df.Data.std()
lb = df.Data.mean()-3*df.Data.std()
print(ub)
print(lb)
```

```python
df_new = df[(df.Data>lb)&(df.Data<ub)] # df>lb is also fine since only one column is present in df
df_new
```

-566 got removed

3. Using Zscore

```python
l = [11,10,-38,12,14,19,260,-566,100,156,23,-97,56,89,189,-10]
#converting list in to dataframe
df = pd.DataFrame(l,columns = ["Data"]) #data is the name of the column
df
```

```python
df = df.sort_values(by = "Data") # by = "which value is using for sorting"
df
```

```python
df["ZScore"] = (df.Data-df.Data.mean())/df.Data.std()
df.ZScore
```

```python
df_new = df[(df.ZScore>-3)&(df.ZScore<3)]
df_new
```

```python
df_new.describe()
```
