```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/Datasets_for_AIML/ML/Data.csv")
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

Extracting dependant and independant variables(X,y)

```python
X = df.iloc[:,:-1].values
X
```

```python
y = df.iloc[:,3].values
y
```

Applying LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,0] = le.fit_transform(X[:,0])
X
```

Creating dummy variables

```python
dummies = pd.get_dummies(df.Country, drop_first = True)
dummies
```

merging the dummy variables in to dataframe

```python
merged = pd.concat([df, dummies],axis = 1)
merged
```

Droping unwanted columns

```python
X = merged.drop(["Country","Purchased"],axis = 1)
X
```

Removing NaN values

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan, strategy = "mean")
X.iloc[:,0:2] = imp.fit_transform(X.iloc[:,0:2])
X
```

Converting y in to numbers

```python
y = le.fit_transform(y)
y
```

Test train split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

Feature scaling (Normalising values to a certain range (-1 to 1))

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train.iloc[:,0:2] = ss.fit_transform(X_train.iloc[:,0:2])
X_train
```
